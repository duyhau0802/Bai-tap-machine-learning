### **Bước 2**:

```python
Mean = numpy.mean(Points, axis=0)   #tính mean
Points_Norm = Points - Mean         #x-mean
#plt.plot(Points_Norm[:, 0], Points_Norm[:, 1], 'bx')
#plt.show()
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(Points_Norm[:, 0], Points_Norm[:, 1], Points_Norm[:, 2])
```

```python
row, col = Points_Norm.shape
Sample_Cov_Matrix = numpy.matmul(Points_Norm.T, Points_Norm)/(row - 1) # ma trận hiệp phương sai
Eigen_Values, Eigen_Vectors = numpy.linalg.eig(Sample_Cov_Matrix)   # tinh giá trị riêng, vector riêng của ma trận hiệp phương sai
```
